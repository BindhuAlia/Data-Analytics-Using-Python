```python
import matplotlib.pyplot as p
```

```python
#plot
x=[3,5]
y=[2,5]
p.plot(x,y)
```

```python
#to pass two lists as x and y arguments
p.plot([2,3,7,5,8,3,5],[2,3,4,7,8,3,5])
```

```python
#lineplot
import numpy as np
t=np.arange(0,5)
a=np.arange(1,6)
p.plot(t,a)
#to show text near a particular point
p.text(2,2.9,' mid point')
```

```python
#to give labels to both axis
p.plot([1,4],[2,7])
p.xlabel('time')
p.ylabel('distance')
p.title('time vs distance')
# to give title in left/right corner
#p.title('abc',loc='left/right')
```

```python
#to change appearance of the line
p.plot([1,2,3,4],[2,3,4,5],'go')
```

```python
p.plot([1,2,3,4],[2,3,4,5],'gd')
```

```python
p.plot([1,2,3,4],[2,3,4,5],'--')#dashed line with default color
```

```python
p.plot([1,2,3,4],[2,3,4,5],'cd')#cyan color
```

```python
help(p.plot)
```

```python
p.plot([1,2,3,4],[2,3,4,5],'g^')
```

```python
p.plot([1,2,3,4],[2,3,4,5],'c',marker='*',mec='black',ms=10,mfc='pink')#marker ex : . , * + h H | _
#mec=color of shape,mfc=color inside shape,ms=marker size
p.grid()
```

```python
a=[1,2,3,4]
b=[2,3,4,5]
p.plot([2,4,6,8],[1,3,5,7],'c',marker='*',ms=10,mec='black',mfc='black')
p.plot(a,b,marker='.',ms=13,mec='black',mfc='black')
p.legend(['Series','Even num'],loc='lower right')# to get names of 2 lines 
p.grid()
```

```python
#barchart:comparision between categorical values(on y axis) and categories(on x axis)
#syntax bar(x,y)
```

```python
a=[1,2,3]
b=['a','b','c']
p.barh(a,b)
```

```python
prog=['c','cpp','java','python','r']
perc=[56,67,78,98,93]
p.bar(prog,perc,color=['pink','black'])
p.title('Programmming Languages')
p.xlabel('languages')
p.ylabel('percentages')
```

```python
#histogram -represents how the frequency of the values 
b=[99,87,78,76,89,88,99,100]
p.hist(b)
```

```python
c=['a','a','a','b','c','c']
p.hist(c,color='black')
```

```python
#scatter plot -- graph will be represented in lineplot by given markers
x=[2,3,4,5]
y=[6,7,8,9]
p.scatter(x,y,marker='*',color='black')
```

```python
#pie chart
prog=['c','cpp','java','python','r']
perc=[56,67,48,98,93]
p.pie(perc,labels=prog,)
p.legend(prog,loc='lower right')
#to save the figure
p.savefig('piechart example.jpg')
```
